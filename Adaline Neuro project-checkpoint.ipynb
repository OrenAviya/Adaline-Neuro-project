### Starting the algorithm

```python
import numpy as np

class Adaline:
    
    def __init__(self, lr=0.01, n_epochs=50, random_state=None):
        self.lr = lr
        self.n_epochs = n_epochs
        self.random_state = random_state
        self.weights = None
        self.bias = None
        self.cost_ = []
     
    #The fit method is responsible for training the Adaline model.
    #It initializes the weights and bias, and then iteratively adjusts them using the input features X and target labels y.
    #The adjustments are made based on the difference between the predicted outputs and the true labels,
    #multiplied by the learning rate lr.
    #The cost (mean squared error) is also calculated and stored in self.cost_ at each iteration.   
    def fit(self, X, y):
        if self.random_state:
            np.random.seed(self.random_state)
            
        # Initialize weights and bias
        self.weights = np.random.randn(X.shape[1])
        self.bias = np.random.randn()
        
        # Fit Adaline
        for i in range(self.n_epochs):
            error = y - self.predict(X)
            self.weights += self.lr * X.T.dot(error)
            self.bias += self.lr * error.sum()
            cost = (error**2).sum() / 2.0
            self.cost_.append(cost)
            
    #The predict method takes the input features X 
    #and returns the predicted labels by comparing the net input to a threshold (0 in this case).       
    def predict(self, X):
        return np.where(self.net_input(X) >= 0, 1, -1)
    
    #The net_input method calculates the net input by multiplying the input features X with the learned weights
    #and adding the bias term.
    def net_input(self, X):
        return X.dot(self.weights) + self.bias
```

```python

# Extract the features and labels
X = data.iloc[:, 1:].values
y = np.where(data.iloc[:, 0].values !=2, -1, 1)

# Split the data into train and test sets
train_ratio = 0.8
num_train = int(train_ratio * len(X))
X_train, X_test = X[:num_train], X[num_train:]
y_train, y_test = y[:num_train], y[num_train:]

# Initialize and fit the Adaline classifier
ada = Adaline(lr=0.0001, n_epochs=1000, random_state=1)
ada.fit(X_train, y_train)

# Predict the labels for the test set
y_pred = ada.predict(X_test)

# Evaluate the accuracy of the classifier
accuracy = (y_pred == y_test).mean()
print(f'Accuracy: {accuracy:.2f}')
std_dev = np.std(y_pred)
print(f'Standard Deviation: {std_dev}')
```

```python
import matplotlib.pyplot as plt

# Initialize and fit the Adaline classifier
ada = Adaline(lr=0.0001, n_epochs=1000, random_state=1)
ada.fit(X_train, y_train)

# Plot the cost function over epochs
plt.plot(range(1, len(ada.cost_) + 1), ada.cost_)
plt.xlabel('Epochs')
plt.ylabel('Cost')
plt.show()
```
